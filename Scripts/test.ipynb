```python
import pandas as pd
```

```python
filepath = 'raw_data/'
```

```python
counties = pd.read_csv(filepath+'us_county_congressionalvote_2014.csv', header=0)
counties.head()
```

```python
counties.drop(columns=['id', 'state_fips', 'county_fips', 'class_fips'], inplace=True)
counties.head()
```

```python
census = pd.read_csv(filepath+'county_2017-2021.csv', header=0)
census.head()
```

```python
census_consolidated = pd.merge(counties, census, how='left', left_on=['fips'], right_on=['geoid_2010'])
```

```python
census_consolidated.shape
```

```python
census_consolidated.drop(columns=['geoid_2010','year'], inplace=True)

census_consolidated.head()
```

```python
grants = pd.read_csv(filepath+'Complete_county_level_grants.csv', header=0)
grants.head()
```

```python
grants.drop(columns=['Population2022',	'Poverty Estimate',	'Poverty Percent',	'Median Household Income'], inplace=True)
grants.head()
```

```python
grants.shape
```

```python
census_consolidated['county_name'] = census_consolidated['county_name'].str.lower().str.replace(r'\s+county$', '', regex=True).str.strip()
census_consolidated.head()
```

```python
census_grants = pd.merge(grants, census_consolidated, how='left', left_on=['county', 'ST'], right_on=['county_name', 'state_code'])
census_grants.head()
```

```python
census_grants.shape
```

```python
census_grants.drop(columns=['county_name', 'state_code'], inplace=True)
census_grants['county'] = census_grants['county'].str.capitalize()
census_grants.rename(columns={'county':'county_name', 'ST':'state_code', 'STATE':'state'}, inplace=True)
census_grants.head()
```

```python
census_grants.to_csv(filepath + 'census-grants.csv', index=False)
```
